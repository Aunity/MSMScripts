```python
import os
import sys
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from msmbuilder.dataset import dataset 
import pyemma
```

```python
jobdir = '/home/songke/maohua/MY1/charmm27'
```

```python
ticas = dataset(jobdir+'/03.tICA/ticas')
```

```python
txx = np.concatenate(ticas)
dtrajs = pd.read_pickle(jobdir+'/firstmodel/05.cluster/n2000/dtrajs.pkl3')
msm = pyemma.msm.bayesian_markov_model(dtrajs, lag=1500, show_progress=False)
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(12,5))
ax = axs[0]
pyemma.plots.plot_free_energy(*txx.T, ax =ax)
ax.set_xlabel('IC 1')
ax.set_ylabel('IC 2')
ax.set_title('MD simulation')

ax = axs[1]
pyemma.plots.plot_free_energy(*txx.T, ax =ax, weights=np.concatenate(msm.trajectory_weights()))
ax.set_xlabel('IC 1')
ax.set_ylabel('IC 2')
ax.set_title('MSM ensemble')
```
