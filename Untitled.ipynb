```python
import os
import sys
import numpy as np
import pandas as pd

import pyemma

import matplotlib as mpl
mpl.style.use('pcca20/ymh.mplstyle')
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

from msmbuilder.dataset import dataset
from msmbuilder.msm import implied_timescales

```

```python
lags = [1]
lags.extend(np.arange(200, 5001,200))
```

```python
seq = pd.read_pickle('pccaplus7/macroseq.pkl3')
```

```python
T = 4
its = implied_timescales(seq, lag_times=lags, n_timescales=10, n_jobs=T, verbose=1)
```

```python
dt = 0.02
fig, ax = plt.subplots()
ax.plot(np.array(lags)*dt, its*dt, linewidth=3)
ax.semilogy()
ax.set_ylim(0,1e4)
ax.set_xlabel('lagtimes (ns)')
ax.set_ylabel('implied timescale (ns)')
fig.tight_layout()
#fig.savefig('timescale.png')
```

```python
macrocenters = []

for i in Ma.state_labels_:
    tmp = []
    for j in  np.array(pccaplus.state_labels_)[pccaplus.microstate_mapping_==int(i)]:
        #print(j)
        tmp.append(centers[j])
    macrocenters.append(np.array(tmp).mean(axis=0))
```
